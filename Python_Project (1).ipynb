#California House dataset




Introduction: The dataset gives us brief description about the california house dataset.The main aim of this datasets is to predict the median house prices of california.In this dataset,we have information regarding the demography(income,population,house occupancy), the location (latitude,longitude) and general information regarding the house (number of rooms,number of bedrooms,age of the house).

Features:
1. Longitude:Represents the longitude of the location and is
   continuous numerical data.
2. Latitude represents the latitude of the location.
3. Housing_median_age:Median age of a house within a block; a lower
   number is a newer building.Statistically,these are discrete values.
3. Total_rooms :It represents total number of rooms in the house.
   Statistically,it is quantitative in nature(discrete data).
4. Total_bedrooms:Total number of bedrooms within a block.It is
   discrete data.
5. Population: Total number of people residing within a block and is
   quantitative (discrete) data.
6.  Households: Total number of households,a group of people residing
   within a home unit, for a block. It is discrete in nature.
7.  MedianIncome: Median income for households within a block of        
    houses  and  data is continuous.
8. MedianHouseValue: Median house value for households within a block
   (discrete).
9. OceanProximity: Location of the house w.r.t ocean/sea.Statistically,
   It is categorical data.

10.Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms
is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.


```python
data=[
    (df['total_bedrooms']<=10),                                   #condition list
    (df['total_bedrooms']>=11) & (df['total_bedrooms']<1000),
    (df['total_bedrooms']>=1000)
     ]
values=['small','medium','large']  #choice list
df['total_bedroom_size']=np.select(data,values) #select function return an array drawn from elements in choicelist,depending o
df
```

The above code added a new column ,total_bedroom_size, which is filled
with information small,medium or large by comparing the
total_bedrooms with the mentioned conditions.
